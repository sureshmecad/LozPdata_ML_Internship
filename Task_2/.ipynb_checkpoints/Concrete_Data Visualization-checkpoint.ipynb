### Importing Libraries

```python
# Read Data
import numpy as np
import pandas as pd

# Visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=(10,8)

# style
plt.style.use("fivethirtyeight")
sns.set_style("darkgrid")
```

### Read Data

```python
# Import first 5 rows
df = pd.read_csv("concrete.csv")
df.head()
```

```python
# shape of data ie Rows x Columns
df.shape
```

```python
df.info()
```

### Missing Data

```python
# checking for null values
df.isnull().sum()
```

```python
df.describe()
```

### Data Visualization

#### Univariate Analysis

```python
# Histrogram
df.hist(figsize = (10,8))
plt.show()
```

```python
plt.hist(df.strength)
plt.show()
```

- Target variable 'strength' is right skewed and imbalanced data.

```python
data = df['water'].head(10)
data.plot.bar()

plt.xlabel('age', fontsize = 16)
plt.ylabel('count', fontsize = 16)

plt.title('age Vs count', fontsize = 25, fontweight = 'bold')
```

```python
x = df['cement'].head(10)
y = df['strength'].head(10)
data.plot.bar(x,y)
```

```python
plt.figure(figsize=(10,8))

x = df['age'].head(30)
data = df['water'].head(30)

plt.bar(x,data);

#plt.legend()

plt.xlabel('age', fontsize = 16)
plt.ylabel('water', fontsize = 16)

plt.title('age Vs water', fontsize = 25, fontweight = 'bold')

plt.xticks(rotation = 45, fontsize = 11)
plt.yticks(fontsize = 11)

plt.show()
```

#### Bivariate Analysis

```python
# Scatter plot
plt.figure(figsize=(10,8))
plt.scatter(df['cement'], df['fineagg'])

plt.xlabel('cement')
plt.ylabel('fineagg')

plt.title('cement Vs fineagg')
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['cement'],df['water'])

plt.xlabel('cement')
plt.ylabel('water')

plt.title('cement Vs water')
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['cement'],df['ash'])

plt.xlabel('cement', fontsize=18)
plt.ylabel('ash', fontsize=18)

plt.title('cement Vs ash')
```

```python
plt.figure(figsize=(9,7))
x = df['age']
y = df['strength']
plt.scatter(df['age'], df['strength'], color='green', linewidth=5, linestyle='dotted')

plt.xlabel('age', fontsize=18)
plt.ylabel('strength', fontsize=18)

plt.title('Age Vs Strength', fontsize=22, fontweight='bold')
```

```python
# Box Plot
plt.figure(figsize=(15,10))
plt.boxplot([df['cement'],df['water'],df['slag'],df['age'],df['strength'],df['ash'],df['superplastic'],
                                                 df['coarseagg'],df['fineagg']])
plt.show()
```

- Observe that we have outliers in water, slag, age, strength, superplastic, fineagg

<h2 style="color:green" align="left"> Data Visualization : Seaborn </h2>

### Univariate Analysis

#### Histogram

```python
sns.distplot(df.strength, kde=True)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

#### Count Plot

```python
plt.figure(figsize=(9,7))
sns.countplot('age', data=df)
```

### Missing Values

```python
# Missing value representation by Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
plt.figure(figsize=(9,7))
sns.heatmap(df.corr(), annot=True)
```

### Pair Plot
- to represent relation between each feature

```python
sns.pairplot(df)
```

### Box Plot
- to find out outliers

```python
plt.figure(figsize=(5,4))
sns.boxplot(data=df.cement, palette='winter')
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(data=df.strength, palette='winter')
```

### Relational Plot

```python
sns.relplot(x='cement', y='strength', data=df, hue="age")
```

### Strip Plot

```python
sns.stripplot(x='cement', y='strength', data=df)
```

### Voilin Plot

```python
fig,ax = plt.subplots(nrows=4, ncols=2, figsize=(10,8))
plt.suptitle('Violin Plots', fontsize=20, fontweight = 'bold')
sns.violinplot(x="cement", data=df, ax=ax[0,0], palette='Set3')
sns.violinplot(x="slag", data=df, ax=ax[0,1], palette='Set3')
sns.violinplot (x='ash', data=df, ax=ax[1,0], palette='Set3')
sns.violinplot(x='water', data=df, ax=ax[1,1], palette='Set3')
sns.violinplot(x="superplastic", data=df, ax=ax[2,0], palette='Set3')
sns.violinplot(x="coarseagg", data=df, ax=ax[2,1], palette='Set3')
sns.violinplot (x ='fineagg', data=df, ax=ax[3,0], palette='Set3')
sns.violinplot(x='age', data=df, ax=ax[3,1], palette='Set3')
plt.show()
```

### Swarm Plot

```python
sns.swarmplot(x='cement', y='strength', data=df)
```

### Regression Plot

```python
sns.lmplot(x='cement', y='strength', data=df)
```

### Dist Plot

```python
fig, ax = plt.subplots(4,2, figsize=(12,10))
sns.distplot(df.cement, bins = 20, ax=ax[0,0]) 
sns.distplot(df.slag, bins = 20, ax=ax[0,1]) 
sns.distplot(df.ash, bins = 20, ax=ax[1,0]) 
sns.distplot(df.water, bins = 20, ax=ax[1,1])
sns.distplot(df.superplastic, bins = 20, ax=ax[2,0]) 
sns.distplot(df.coarseagg, bins = 20, ax=ax[2,1]) 
sns.distplot(df.fineagg, bins = 20, ax=ax[3,0]) 
sns.distplot(df.age, bins = 20, ax=ax[3,1])
plt.show()
```
