# $$Introduction$$

## Concrete Compresive strength analysis and Prediction Model
### 1. Identify the problem
Concrete is the most important material in civil engineering. The concrete compressive strength is a highly nonlinear function of age and ingredients. 

### 2. Objective
I develop a Machine Learning Model which is predict Concrete compressive strength.

### 3. Data Source
Number of instances 1030

Number of Attributes 9

Attribute breakdown 8 quantitative input variables, and 1 quantitative output variable
Missing Attribute Values None.
[Data Available](https://archive.ics.uci.edu/ml/machine-learning-databases/concrete/compressive/)

######  Concrete compressive strength: 
The compressive strength of a concrete is defined as the ability of the concrete to withstand specific compressive forces.
# $$  Strength = \frac{Force}{Area} $$

## Attribute Information:
Cement (component 1) quantitative -- kg in a m3 mixture

Blast Furnace Slag (component 2) quantitative -- kg in a m3 mixture

Fly Ash (component 3) -- quantitative -- kg in a m3 mixture

Water (component 4) -- quantitative -- kg in a m3 mixture

Superplasticizer (component 5) -- quantitative -- kg in a m3 mixture

Coarse Aggregate (component 6) -- quantitative -- kg in a m3 mixture

Fine Aggregate (component 7) -- quantitative -- kg in a m3 mixture

Age -- quantitative -- Day (1~365)

Concrete compressive strength -- quantitative

### Dist Plot

```python
fig, ax = plt.subplots(4,2, figsize=(10,8))
sns.distplot(data.cement, bins = 20, ax=ax[0,0]) 
sns.distplot(data.slag, bins = 20, ax=ax[0,1]) 
sns.distplot(data.ash, bins = 20, ax=ax[1,0]) 
sns.distplot(data.water, bins = 20, ax=ax[1,1])
sns.distplot(data.superplastic, bins = 20, ax=ax[2,0]) 
sns.distplot(data.coarseagg, bins = 20, ax=ax[2,1]) 
sns.distplot(data.fineagg, bins = 20, ax=ax[3,0]) 
sns.distplot(data.age, bins = 20, ax=ax[3,1])
plt.show()
```

### Violin plot

```python
fig,ax = plt.subplots(nrows=4, ncols=2, figsize=(10,8))
plt.suptitle('Violin Plots',fontsize=24)
sns.violinplot(x="cement", data=data,ax=ax[0,0],palette='Set3')
sns.violinplot(x="slag", data=data,ax=ax[0,1],palette='Set3')
sns.violinplot (x ='ash', data=data, ax=ax[1,0], palette='Set3')
sns.violinplot(x='water', data=data, ax=ax[1,1],palette='Set3')
sns.violinplot(x="superplastic", data=data,ax=ax[2,0],palette='Set3')
sns.violinplot(x="coarseagg", data=data,ax=ax[2,1],palette='Set3')
sns.violinplot (x ='fineagg', data=data, ax=ax[3,0], palette='Set3')
sns.violinplot(x='age', data=data, ax=ax[3,1],palette='Set3')
plt.show()
```

### Correlation
